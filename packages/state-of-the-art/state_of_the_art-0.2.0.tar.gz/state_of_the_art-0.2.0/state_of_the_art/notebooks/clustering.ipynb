```python
# get latest period papers

import datetime
from state_of_the_art.paper.papers_data_loader import PapersLoader
# read april of 2025
papers_df = PapersLoader().load_between_dates(start=datetime.date(2025, 4, 1), end=datetime.date(2025, 4, 30))

len(papers_df)
papers_df.head()
```

```python

# convert title and abstract to text
papers_df["text"] = papers_df["title"] + " " + papers_df["abstract"]
text_list = papers_df["title"].tolist()

papers_df['text']
# convert text to semantic embeddings
from sentence_transformers import SentenceTransformer

## use a low dimension model for clustering  
model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(papers_df["text"].tolist())

embeddings[0]
```

```python
papers_df["embedding"] = embeddings.tolist()
```

```python

import numpy as np
from sklearn.cluster import AgglomerativeClustering

NUM_OF_CLUSTERS = 3
clustering = AgglomerativeClustering(n_clusters=NUM_OF_CLUSTERS)
# convert embeddings to numpy array
clustering.fit(embeddings)




#papers_df["cluster"] = kmeans.predict(embeddings)
```

```python
max_papers_per_cluster = 7 

for cluster_index in range(NUM_OF_CLUSTERS):
    class_i_papers = [] 

    for i in range(len(text_list)):
        if clustering.labels_[i] == cluster_index:
            class_i_papers.append( f"{i} - {text_list[i]}")  



    print(f"Class {cluster_index} papers:  ")
    printed = 0
    for paper in class_i_papers:
        print(paper)
        printed += 1
        if printed >= max_papers_per_cluster:
            break
    print("\n\n")
```
