```python

import datetime
from state_of_the_art.tables.paper_table import PaperTable

date_from = datetime.datetime.strptime('2024-01-01', '%Y-%m-%d').date()
date_to = datetime.datetime.strptime('2025-03-31', '%Y-%m-%d').date()

papers_df = PaperTable().read()
papers_df = papers_df[(papers_df["published"].dt.date >= date_from) & (papers_df["published"].dt.date <= date_to)]

# put title in the first column, keep the rest of the columns in the same order do not remove any columns   
papers_df = papers_df[["title"] + [col for col in papers_df.columns if col != "title"]]

papers_df.describe()
```

```python
from state_of_the_art.relevance_model.text_evaluation_inference import PersonalPreferenceInference


result = PersonalPreferenceInference().predict_batch(papers_df['title'].to_list())
```

```python
papers_df['relevance_prediction'] = result
```

```python
from notebooks.deep_recommender.scores_table import ScoresTable
import datetime
import numpy as np

table = ScoresTable()

current_data: pd.DataFrame = table.read()
# convert all strings in bm25_sum_score to float or NAN
current_data.describe()
```

```python
final_result = papers_df[['title', 'abstract_url', 'relevance_prediction']].copy()
# renaem title to paper_title
final_result.rename(columns={'title': 'paper_title'}, inplace=True)
# rename abstract_url to paper_url
final_result.rename(columns={'abstract_url': 'paper_url'}, inplace=True)
# rename relevance_prediction to predictor_score
final_result.rename(columns={'relevance_prediction': 'predictor_score'}, inplace=True)



result_df = pd.merge(current_data, final_result, on='paper_url', how='left')
# drop paper_title_x
result_df = result_df.drop(columns=['paper_title_x'])
# drop all _x and _y suffixes from column names
result_df.columns = result_df.columns.str.replace('_x', '').str.replace('_y', '')

result_df

```

```python
# final score is bm25_sum_score * predictor_score


result_df['final_score'] = result_df['bm25_sum_score'] * result_df['predictor_score']
result_df

# sort by final_score descending
result_df = result_df.sort_values(by='final_score', ascending=False)
# see from 50 to 150
result_df[['paper_title', 'final_score','paper_url']].iloc[150:200]
```
